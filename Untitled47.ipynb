#### 2. Advanced NLP Logic for Query Processing
To handle these complex queries, the chatbot will need a more sophisticated NLP pipeline:

*   **Intent Recognition:**
    *   **Method:** Utilize a classification model (e.g., using a pre-trained BERT-like model fine-tuned on housing-related intents or a simpler SVM with TF-IDF features). This model would classify the user's query into predefined intents such as: `FIND_PROPERTY`, `COMPARE_AREAS`, `GET_STATISTICS`, `GET_RELATIONSHIP`, `SUMMARIZE_DESCRIPTION`, `FOLLOW_UP_QUESTION`, `UNKNOWN_INTENT`.
    *   **Example:**
        *   "Show me properties..." -> `FIND_PROPERTY`
        *   "Are properties...older or newer?" -> `COMPARE_AREAS`
        *   "Is there a correlation...?" -> `GET_RELATIONSHIP`
        *   "What's the typical number...?" -> `GET_STATISTICS`

*   **Entity Extraction (Named Entity Recognition - NER):**
    *   **Method:** Employ a sequence labeling model (e.g., Bi-LSTM-CRF, SpanBERT, or even rule-based regex for numerical values). This will identify and label key entities in the query.
    *   **Entities to Extract:**
        *   `FEATURE`: (e.g., 'median income', 'house value', 'housing age', 'rooms', 'bedrooms per room', 'population per household').
        *   `VALUE`: (e.g., 'high', 'low', 'young', 'large', numerical values like '5.00', '3000', '0.2').
        *   `OPERATOR`: (e.g., 'greater than', 'less than', 'between', 'around', 'equal to', 'typically', 'usually').
        *   `QUANTIFIER`: (e.g., 'how many', 'list', 'show me', 'summarize').
        *   `COMPARISON_TERM`: (e.g., 'older', 'newer', 'more', 'fewer', 'affect', 'correlation').
    *   **Mapping:** Link extracted entities to DataFrame column names (e.g., 'median income' -> `median_income_category` or `median_income`).

*   **Relationship Extraction:**
    *   **Method:** Use dependency parsing or rule-based patterns to understand how entities relate to each other and to the overall intent. For comparative queries, identify the two entities being compared and the comparison metric. For relationship queries, identify the two features for which a relationship is sought.
    *   **Example:** In "high income, young housing, and a large number of rooms," the NLP should identify `income=high`, `housing_age=young`, `rooms=large` as conjunctive conditions.

*   **Contextual Understanding (for Follow-up Questions):**
    *   **Method:** Maintain a short-term memory of the previous turn's extracted intent, entities, and filters. If a `FOLLOW_UP_QUESTION` intent is detected, refer back to the last set of applied filters or the subject of the previous query.
    *   **Example:** If the previous query filtered for "high income" properties, a follow-up "And how old are they usually?" would automatically apply the "high income" filter to the age query.

#### 4. Method for Translating to Structured Queries
Once the intent, entities, and relationships are extracted, they need to be translated into executable Pandas DataFrame queries:

1.  **Intent-Based Query Structure:** Each intent will have a corresponding Python function or template that constructs the Pandas query.
    *   `FIND_PROPERTY`: Generates `df[(condition1) & (condition2)]` and returns sample rows.
    *   `GET_STATISTICS`: Generates `df[condition].describe()` or `df[condition].mean()`.
    *   `COMPARE_AREAS`: Performs two separate filtered queries and compares a specific metric.
    *   `GET_RELATIONSHIP`: Triggers a correlation calculation (e.g., `df[features].corr()`) or a visualization function.

2.  **Filter Construction:**
    *   **Categorical:** Entities like `('median_income_category', 'High')` translate to `df['median_income_category'] == 'High'`.
    *   **Numerical:** Entities like `('total_rooms', 'greater than', 3000)` translate to `df['total_rooms'] > 3000`.
    *   **Range:** `('housing_median_age', 'between', 20, 40)` translates to `(df['housing_median_age'] >= 20) & (df['housing_median_age'] <= 40)`.
    *   **Approximate:** `('median_income', 'around', 5.0)` translates to `(df['median_income'] >= 5.0 * 0.9) & (df['median_income'] <= 5.0 * 1.1)`.

3.  **Dynamic Query Assembly:** A Python function will take the extracted intent and a dictionary of extracted `FEATURE: (OPERATOR, VALUE)` pairs, and dynamically build the Pandas filtering and aggregation logic.
    *   For example, if intent is `FIND_PROPERTY` and entities are `{'median_income': ('==', 'High'), 'housing_median_age': ('>', 30)}`, it would construct: `df[ (df['median_income_category'] == 'High') & (df['housing_median_age'] > 30) ]`.

## Develop a Flexible Data Access and Reasoning Engine

Design a system that can dynamically access and reason over the housing data based on the interpreted user queries. This will involve translating complex NLP interpretations into efficient queries against the Pandas DataFrame, potentially combining multiple filters and aggregations to find answers.


#### 1. Define the `execute_query` function structure

```python
def execute_query(intent, entities, dataframe):
    """
    Executes a query against the housing data based on the interpreted intent and entities.

    Args:
        intent (str): The recognized intent (e.g., 'FIND_PROPERTY', 'GET_STATISTICS',
                      'COMPARE_AREAS', 'GET_RELATIONSHIP').
        entities (dict): A dictionary of extracted entities (e.g., {'feature': 'median_income',
                                   'operator': 'greater_than', 'value': 5.0,
                                   'category': {'median_house_value_category': 'High'}}).
        dataframe (pd.DataFrame): The target Pandas DataFrame (df or df_test).

    Returns:
        Union[pd.DataFrame, dict, str]: The query results in a structured format, or an error message.
    """
    filtered_df = dataframe.copy()
    query_results = {}

    # --- Helper function to apply filters ---
    def apply_filters(df, filters):
        temp_df = df.copy()
        for filter_type, filter_data in filters.items():
            if filter_type == 'categorical':
                for col, value in filter_data.items():
                    if col in temp_df.columns and value is not None:
                        temp_df = temp_df[temp_df[col] == value]
            elif filter_type == 'numerical':
                for feature_col, conditions in filter_data.items():
                    if feature_col in temp_df.columns:
                        for condition_type, value_or_range in conditions.items():
                            if condition_type == 'greater_than':
                                temp_df = temp_df[temp_df[feature_col] > value_or_range]
                            elif condition_type == 'less_than':
                                temp_df = temp_df[temp_df[feature_col] < value_or_range]
                            elif condition_type == 'equal_to':
                                temp_df = temp_df[temp_df[feature_col] == value_or_range]
                            elif condition_type == 'between':
                                lower, upper = value_or_range
                                temp_df = temp_df[(temp_df[feature_col] >= lower) & (temp_df[feature_col] <= upper)]
                            elif condition_type == 'around':
                                value, epsilon = value_or_range
                                temp_df = temp_df[(temp_df[feature_col] >= value - epsilon) & (temp_df[feature_col] <= value + epsilon)]
        return temp_df

    # Extract filters from entities
    # This part needs sophisticated NLP parsing to fill the 'filters' dict
    # For this design, we will assume entities is structured like:
    # entities = {
    #     'filters': {
    #         'categorical': {'median_house_value_category': 'High', 'median_income_category': 'Medium'},
    #         'numerical': {'total_rooms': {'greater_than': 2000}, 'housing_median_age': {'between': [20, 40]}}
    #     },
    #     'statistics_features': ['median_house_value', 'median_income'], # for GET_STATISTICS
    #     'comparison_criteria': {'feature': 'housing_median_age', 'group1_filters': {}, 'group2_filters': {}}, # for COMPARE_AREAS
    #     'relationship_features': ['median_income', 'median_house_value'] # for GET_RELATIONSHIP
    # }
    query_filters = entities.get('filters', {})
    filtered_df = apply_filters(filtered_df, query_filters)

    if filtered_df.empty:
        return "No records found matching the specified criteria."

    if intent == 'FIND_PROPERTY':
        # Return a sample of the filtered DataFrame or the count
        num_samples = entities.get('num_samples', 5)
        if 'count_only' in entities and entities['count_only']:
            return {'count': len(filtered_df)}
        else:
            # Select key columns for display
            display_cols = ['longitude', 'latitude', 'median_house_value', 'median_income', 'housing_median_age', 'rooms_per_household']
            actual_display_cols = [col for col in display_cols if col in filtered_df.columns]
            return filtered_df[actual_display_cols].head(num_samples).to_dict('records')

    elif intent == 'GET_STATISTICS':
        statistics_features = entities.get('statistics_features', [])
        if not statistics_features:
            return "Please specify features for which to get statistics."

        stats_result = {}
        for feature in statistics_features:
            if feature in filtered_df.columns:
                stats_result[feature] = {
                    'mean': filtered_df[feature].mean(),
                    'median': filtered_df[feature].median(),
                    'std': filtered_df[feature].std(),
                    'min': filtered_df[feature].min(),
                    'max': filtered_df[feature].max()
                }
            else:
                stats_result[feature] = "Feature not found."
        return stats_result

    elif intent == 'COMPARE_AREAS':
        comparison_criteria = entities.get('comparison_criteria', {})
        if not comparison_criteria or 'feature' not in comparison_criteria:
            return "Please specify comparison criteria and feature."

        feature_to_compare = comparison_criteria['feature']
        group1_filters = comparison_criteria.get('group1_filters', {})
        group2_filters = comparison_criteria.get('group2_filters', {})

        df_group1 = apply_filters(dataframe.copy(), group1_filters)
        df_group2 = apply_filters(dataframe.copy(), group2_filters)

        if df_group1.empty or df_group2.empty:
            return "One or both comparison groups yielded no data."

        if feature_to_compare not in dataframe.columns:
            return f"Feature '{feature_to_compare}' not found for comparison."

        mean1 = df_group1[feature_to_compare].mean()
        mean2 = df_group2[feature_to_compare].mean()

        return {
            'comparison_feature': feature_to_compare,
            'group1_mean': mean1,
            'group2_mean': mean2,
            'difference': mean1 - mean2
        }

    elif intent == 'GET_RELATIONSHIP':
        relationship_features = entities.get('relationship_features', [])
        if len(relationship_features) != 2:
```

````python
import re
import numpy as np # Make sure numpy is imported for np.abs

def process_user_query(user_query, dataframe):
    user_query_lower = user_query.lower()
    response_text = "I'm sorry, I couldn't understand your request. Please try rephrasing your question or ask about specific housing characteristics like median house value, income, age, total rooms, or their relationships. You can also specify 'low', 'medium', or 'high' categories."

    # --- Feature Mapping for easier lookup (used by intents) ---
    feature_col_map = {
        'total rooms': 'total_rooms', 'total bedrooms': 'total_bedrooms',
        'population': 'population', 'households': 'households',
        'median income': 'median_income', 'income': 'median_income',
        'median house value': 'median_house_value', 'house value': 'median_house_value', 'housing value': 'median_house_value',
        'housing median age': 'housing_median_age', 'housing age': 'housing_median_age', 'age': 'housing_median_age',
        'rooms per household': 'rooms_per_household',
        'bedrooms per room': 'bedrooms_per_room',
        'population per household': 'population_per_household'
    }

    # --- Categorical Filter Keywords ---
    categorical_keywords_map = {
        'median_house_value_category': {'low house value': 'Low', 'medium house value': 'Medium', 'high house value': 'High', 'low prices': 'Low', 'medium prices': 'Medium', 'high prices': 'High'},
        'median_income_category': {'low income': 'Low', 'medium income': 'Medium', 'high income': 'High'},
        'housing_median_age_category': {'new homes': 'New', 'medium age homes': 'Medium', 'old homes': 'Old', 'young housing': 'New'}
    }

    # Helper to parse conditions from a raw string
    def _parse_filters_from_query_text(text):
        filters = {'categorical': {}, 'numerical': {}}
        local_temp_cat_conditions = {col: [] for col in categorical_keywords_map.keys()}

        # Parse categorical conditions
        for category_col, keyword_map in categorical_keywords_map.items():
            for keyword, cat_value in keyword_map.items():
                if keyword in text:
                    local_temp_cat_conditions[category_col].append(cat_value)

        # Apply categorical conditions (and check conflicts)
        for col, values in local_temp_cat_conditions.items():
            if len(values) > 1 and len(set(values)) > 1:
                return None, f"Your query contains conflicting criteria for {col.replace('_category', '').replace('_', ' ')}. Please specify only one category (e.g., 'low', 'medium', or 'high')."
            elif len(values) == 1:
                filters['categorical'][col] = values[0]

        # Parse numerical conditions
        for display_name, col_name in feature_col_map.items():
            # Pattern for 'feature (operator) value'
            match_op_val = re.search(f"\\b{re.escape(display_name)}\\s+(greater than|less than|equal to|around)\s+(\\d+(?:\\.\\d+)?)\\b", text)
            if match_op_val:
                operator_str = match_op_val.group(1)
                value = float(match_op_val.group(2))
                if col_name not in filters['numerical']: filters['numerical'][col_name] = {}
                if operator_str == 'greater than': filters['numerical'][col_name]['greater_than'] = value
                elif operator_str == 'less than': filters['numerical'][col_name]['less_than'] = value
                elif operator_str == 'equal to': filters['numerical'][col_name]['equal_to'] = value
                elif operator_str == 'around': filters['numerical'][col_name]['around'] = (value, value * 0.1)

            # Pattern for 'feature between value1 and value2'
            match_between = re.search(f"\\b{re.escape(display_name)}\\s+between\\s+(\\d+(?:\\.\\d+)?)\s+and\\s+(\\d+(?:\\.\\d+)?)\\b", text)
            if match_between:
                value1 = float(match_between.group(1))
                value2 = float(match_between.group(2))
                if col_name not in filters['numerical']: filters['numerical'][col_name] = {}
                filters['numerical'][col_name]['between'] = ([value1, value2])
        return filters, None

    # --- Intent Determination (Ordered by specificity) ---

    # GET_RELATIONSHIP
    match = re.search(r'(relationship|correlation) between (.+?) and (.+)', user_query_lower)
    if match:
        feature1_name = match.group(2).strip()
        feature2_name = match.group(3).strip()

        feat1_col = feature_col_map.get(feature1_name)
        feat2_col = feature_col_map.get(feature2_name)

        if feat1_col and feat2_col:
            intent = 'GET_RELATIONSHIP'
            entities = {'relationship_features': [feat1_col, feat2_col]}
            return generate_response_from_query_results(intent, entities, execute_query(intent, entities, dataframe))
        else:
            return f"Could not find features '{feature1_name}' or '{feature2_name}' for relationship query."

    # COMPARE_AREAS
    # Refined regex to better capture comparison groups with 'between X and Y' or 'X vs Y'
    comp_feature_match = re.search(r'compare\s+(.+?)(?:\s+between|\s+in areas with)?\s*(.+?)(?:\s+vs|and)(?:\s+the areas with|\s+in areas with)?\s*(.+)', user_query_lower)
    if comp_feature_match:
        feature_to_compare_raw = comp_feature_match.group(1).strip()
        group1_raw = comp_feature_match.group(2).strip()
        group2_raw = comp_feature_match.group(3).strip()

        feature_to_compare = feature_col_map.get(feature_to_compare_raw)
        if not feature_to_compare:
            return f"Could not identify the feature to compare: '{feature_to_compare_raw}'."

        group1_filters, error1 = _parse_filters_from_query_text(group1_raw)
        if error1: return error1
        group2_filters, error2 = _parse_filters_from_query_text(group2_raw)
        if error2: return error2

        intent = 'COMPARE_AREAS'
        entities = {
            'comparison_criteria': {
                'feature': feature_to_compare,
                'group1_filters': group1_filters,
                'group2_filters': group2_filters,
                'group1_description': group1_raw, # Store raw description for response generation
                'group2_description': group2_raw
            }
        }
        return generate_response_from_query_results(intent, entities, execute_query(intent, entities, dataframe))

    # GET_STATISTICS or FIND_PROPERTY (general cases - fallthrough if no specific comparison/relationship)

    # Parse all filters from the entire query string once for these general intents
    global_filters, error = _parse_filters_from_query_text(user_query_lower)
    if error: return error # Return conflict error if detected during global parsing

    # Default entities with global filters
    entities = {'filters': global_filters}

    if 'average' in user_query_lower or 'mean' in user_query_lower or 'typical' in user_query_lower or 'statistics' in user_query_lower:
        intent = 'GET_STATISTICS'
        statistics_features = []
        # Extract features for statistics
        for display_name, col_name in feature_col_map.items():
            if display_name in user_query_lower:
                statistics_features.append(col_name)
        if not statistics_features:
            # Default features if intent is GET_STATISTICS but no specific feature is mentioned
            statistics_features = ['median_house_value', 'median_income', 'housing_median_age'] # Example defaults
        entities['statistics_features'] = statistics_features
        return generate_response_from_query_results(intent, entities, execute_query(intent, entities, dataframe))


    if 'show me' in user_query_lower or 'list' in user_query_lower or 'find properties' in user_query_lower or 'how many' in user_query_lower:
        intent = 'FIND_PROPERTY'
        num_samples_match = re.search(r'(\d+)\s+examples', user_query_lower)
        if num_samples_match: entities['num_samples'] = int(num_samples_match.group(1))

        if 'how many' in user_query_lower: entities['count_only'] = True

        return generate_response_from_query_results(intent, entities, execute_query(intent, entities, dataframe))

    # Fallback for implicit FIND_PROPERTY if filters are present and no other intent matched
    if global_filters['categorical'] or global_filters['numerical']:
        intent = 'FIND_PROPERTY'
        return generate_response_from_query_results(intent, entities, execute_query(intent, entities, dataframe))

    return response_text
````
